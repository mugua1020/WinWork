```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False

price = [3711234,1879215,464589,1654125,504007,57642,0,72068,183617]
label=['memory load','memory store','integer operation','floating ponit operation',
          'conditional branch','unconditional branch','atomic instr','cache instr','other instr']

"""
绘制水平条形图方法barh
参数一：y轴
参数二：x轴
"""
plt.barh(range(9), price, height=0.7, color='steelblue', alpha=0.8)      # 从下往上画
plt.yticks(range(9), label)
#plt.xlim(30,47)
plt.xlabel("指令个数")
plt.title("不同指令个数直方图")
for x, y in enumerate(price):
    plt.text(y + 0.2, x - 0.1, '%s' % y)
plt.show()
```
